```python
import numpy as np
import os
import PIL as Image
import PIL.Image
```

```python
def resize(files, Dir):  
    print("\n\n----------X------------X-----------------------X\n\n")
    #arr = np.zeros([20,200,200,3])
    counter = 0
    pic = 1
   
    for i in files:     # create array equal to pictures quantity
        list = os.listdir(".\photos")
        number_files = len(list)
        arr = np.zeros([number_files,200,200,3])
        
        print("PIC NO:",pic)# Resizing and saving Photos in array
        img = PIL.Image.open(Dir+"//"+i)
        n_img = img.resize((200,200))
        arr[counter] = n_img
        print("Image Resized:")
        print("Shape :",arr.shape)
        print("Size :",arr.size)
        #print(arr)
        print("\n\n----------X------------X-----------------------X\n\n")
        pic+=1
        counter+=1
        
def search():  #Searching for files   
    print("Files found :")
    for Dir, List, files in os.walk("./Photos"):
        for i in files:
            print("\t",i,)
        resize(files, Dir)
```

```python
if __name__== "__main__":
    search()
    
```
