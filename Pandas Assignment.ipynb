### 1. Create a data series with marks of students : 75, 80, 79, 60

```python
import pandas as pd
import os
marks_series = pd.Series([75,80,79,60])
display(marks_series)
type(marks_series)

name_list = ["Bilal","Danyal","Jamal","Maha"]
```

### 2. Create a data frame with name of students, id and marks

```python
dic = {"name":name_list,"id":[1,2,3,4],"marks":marks_series}
data = pd.DataFrame(dic)
display(data)
type(data)
#data.to_csv("data.csv")
```

### 5. Sort the data based on Marks obtained. Fill all the 'na' cells with 0

### 9. Display only those students who got failed in examination

### 10. Find out the basic statistical info about data

```python
print("Mean of all Students :",a["Total (100)"].mean())
print("Median of all Students :",a["Total (100)"].median())
```

### 11. How many students got A, B, C, F?

```python
criteria1 = a["Grade"] == "A"
criteria2 = a["Grade"] == "B"
criteria3 = a["Grade"] == "C"
print("Students got A :",a[criteria1].count().mean())
print("Students got B :",a[criteria2].count().mean())
print("Students got C :",a[criteria3].count().mean())
print("Students got F :",a[criteria].count().mean())
```

### 12. What are the mean scores for students who got A, B, C, F?

```python
A_Average = a[criteria1]
print("Average(mean) of A : ",A_Average["Total (100)"].mean())
B_Average = a[criteria2]
print("Average(mean) of B : ",B_Average["Total (100)"].mean())
C_Average = a[criteria3]
print("Average(mean) of C : ",C_Average["Total (100)"].mean())
F_Average = a[criteria]
print("Average(mean) of F : ",F_Average["Total (100)"].mean())
```
